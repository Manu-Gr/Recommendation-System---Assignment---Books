#### **Colloborative filtering using Correlation Matrix**

```python
cor_piv_table =pd.pivot_table(data = book1, values = 'Book.Rating', index ='Book.Title',columns = 'User.ID').fillna(0)
cor_piv_table
```

```python
# Replacing index values with unique user ID
cor_piv_table.columns = book1['User.ID'].unique()
```

```python
cor_piv_table
```

```python
cor_piv_table.corr()
```

```python
cor1 = cor_piv_table.corr()
```

```python
# fill diagonal with 0
cor2 = cor1.values
cor2
```

#### **Colloborative filtering using Euclidean Matrix**

```python
from sklearn.metrics import pairwise_distances
```

```python
Euclidean_piv_table =pd.pivot_table(data = book1, values = 'Book.Rating', index ='User.ID',columns = 'Book.Title').fillna(0)
Euclidean_piv_table
```

```python
Euclidean_piv_table.index = book1['User.ID'].unique()
```

```python
Euclidean_piv_table
```

```python
Euclidean_piv_table.values
```

```python
Euc1 = pairwise_distances(X = Euclidean_piv_table.values, metric='euclidean')
Euc1
```

```python
Euc2          = pd.DataFrame(data= Euc1)
Euc2.index    = book1['User.ID'].unique()
Euc2.columns  = book1['User.ID'].unique()
Euc2.round(2)
```

#### **Colloborative filtering using Cosine Matrix**

```python
Cosine_piv_table =pd.pivot_table(data = book1, values = 'Book.Rating', index ='User.ID',columns = 'Book.Title').fillna(0)
Cosine_piv_table
```

```python
Cosine_piv_table.index = book1['User.ID'].unique()
```

```python
Cosine_piv_table
```

```python
Cosine_piv_table.values
```

```python
Cosine_ubcf =  1 - pairwise_distances(X = Cosine_piv_table.values, metric='cosine')
Cosine_ubcf
```

```python
Cosine = pd.DataFrame(data= Cosine_ubcf)
Cosine.index    = book1['User.ID'].unique()
Cosine.columns  = book1['User.ID'].unique()
Cosine
```

```python
first_100_users = Cosine.iloc[:100,:100]
first_100_users
```

```python
# fill diagonal with 0
first_100_users = np.fill_diagonal(a = first_100_users.to_numpy(),val = 0)
first_100_users
```

```python
first_100_users.idxmax(axis=1)
```
